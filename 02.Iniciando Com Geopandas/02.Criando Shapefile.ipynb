# Criando Shapefile

## Obtendo os dados

https://geoftp.ibge.gov.br/organizacao_do_territorio/malhas_territoriais/malhas_municipais/municipio_2017/UFs/RJ/

## Lendo

```python
%matplotlib inline
import geopandas as gpd
```

```python
rj = gpd.read_file('../01.Dados/Mapas/RJ/33MUE250GC_SIR.shp')
```

```python
rj
```

```python
rj.plot(color='white', edgeColor='black', figsize=(20,10))
```

## Modificando

```python
rj = rj[rj['NM_MUNICIP'] == 'RIO DE JANEIRO']
```

```python
rj
```

```python
rj.plot(color='orange', edgeColor='black', figsize=(20,10))
```

## Salvando


```python
import os

dir = '../01.Dados/Mapas.RJ-MUNIC'
if not os.path.exists(dir):
    os.makedirs(dir)
```

```python
rj.to_file(dir + '/RJ-MUNIC.shp')
```

```python
rj_munic = gpd.read_file(dir + '/RJ-MUNIC.shp')
```

```python
rj_munic.plot(color='orange', edgeColor='black', figsize=(20,10))
```
